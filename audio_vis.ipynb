```
import os

def plot(file):
    from scipy.io.wavfile import read
    import matplotlib.pyplot as plt

    # read audio samples
    input_data = read(file)
    audio = input_data[1]
    # plot the first 1024 samples
    plt.plot(audio)
    # label the axes
    plt.ylabel("Amplitude")
    plt.xlabel("Time")
    # set the title  
    plt.title("Sample Wav")
    # display the plot
    plt.show()


for x in os.listdir('/content'):
    if x.endswith('.wav'):
        file = os.path.join('/content',x)
        print(file)
        plot(file)
```
